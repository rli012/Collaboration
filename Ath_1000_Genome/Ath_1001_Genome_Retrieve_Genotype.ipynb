```python
################################################################

import h5py, numpy
f = h5py.File('imputed_snps_binary.hdf5','r')
```

```python
Chr = 5
Pos = 15001862
```

```python

# Get all SNP positions for all chromosomes (len=10709949)
positions = f['positions'][:]

# Array of tupels with start/stop indices for each chromosome
chr_regions = f['positions'].attrs['chr_regions']
indices_for_chr1 = chr_regions[Chr-1]

idx_start_pos = indices_for_chr1[0]
idx_end_pos = indices_for_chr1[1]

# Subset positions for SNPs on Chr1
positions_on_chr1 = positions[indices_for_chr1[0]:indices_for_chr1[1]]

# Get SNPs
snps_in_region = f['snps'][idx_start_pos:idx_end_pos]
#snps_in_region = f['snps'][8428147:9638147]
# Find index of a specific position
ix = numpy.where(positions_on_chr1 == Pos)[0][0]

# Get the corresponding SNP for that position
snp = snps_in_region[ix]

## accessions
acc = f['accessions'][:]
```

```python
accessions = []
for a in acc:
    accessions.append(a.decode("utf-8"))
```

```python
fh = open('Chr'+str(Chr)+'_'+str(Pos)+'.txt', 'wt')
fh.write('chr\tpos\t'+'\t'.join(map(str,accessions))+'\n')
```
